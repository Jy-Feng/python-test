```python
#使用matplotlib进行绘图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#入门
# 准备数据
# 创建并配置画布
# 使用数据来绘制不同种类的图形
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
# 显示图像
```

```python
#折线图 表示数据的变化趋势
# 准备数据
list1 = np.arange(1,11)
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.plot(x轴数据集, y轴数据集)
plt.plot(list1,-list1)
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")
# 显示图像
plt.show()
```

```python
#解决绘图中文显示问题
plt.rcParams['font.family']= "simsun" #simhei
#负号使用unicode，中文字体环境下显示不出来
#解决负号显示不出来的问题
plt.rcParams['axes.unicode_minus']=False
```

```python
#创建电影历史票房top5数据
movies =["哥斯拉大战金刚","阿凡达2","流浪地球2", "唐人街探案4", "指环王4"]
moneys = [20, 70, 60,40,30]
#top5电影的前作数据
movies_2=["金刚/哥斯拉大战金刚", "阿凡达/阿凡达2","流浪地球/流浪地球2", "唐人街探案3/唐人街探案4", "指环王3/指环王4"]
moneys_2 = [10,50,50,30,15]
```

```python
#柱状图（垂直，水平）使用高度或长度表示数据，对比各个数据点，查看他们之间的大小关系
# 准备数据
list1 = np.arange(1,11)
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.bar(x轴数据集, height=y轴数据集, width=柱子宽度) align柱子与刻度对齐方式:edge,center
plt.bar(movies_2,height=moneys,width=0.3, color="green",align="edge")
plt.bar(movies_2,moneys_2,width=-0.3,color="blue",align="edge")
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("电影历史票房top5", fontsize=24, color="red")
#配置y轴标签 ylabel  xlabel
plt.ylabel("电影票房(亿人民币)", fontsize=18,color="purple")
#配置y轴的刻度和y轴刻度的标签
plt.yticks([10,30,50,70],labels=["10亿","30亿","50亿","70亿"], fontsize=8)
#配置x轴刻度字体大小颜色
plt.xticks(fontsize=10,color="red")
#配置图例，绘制的图形的说明 loc=图例位置  upper lower left right center 两两组合 正中央center
plt.legend(["top5电影","top电影前作"], loc="upper left")
# 显示图像
plt.show()
# 通用的参数fontsize字体大小，color颜色
```

```python
#水平柱状图 
# 准备数据
list1 = np.arange(1,11)
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.bar(y轴数据集, width=x轴数据集, height=柱子高度)
plt.barh(list1,width=list1,height=0.5)
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")
# 显示图像
plt.show()
```

```python
#散点图 显示数据的分布，要求数据量够大
# 准备数据
list_y = np.random.uniform(0,100,1000)
list_x = np.arange(1000) 
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.scatter(x轴数据集,y轴数据集,s=点大小,c=点颜色)
plt.scatter(list_x,list_y,s=10,c="green")
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")
# 显示图像
plt.show()
```

```python
#饼图  显示分类数据之间的百分占比关系，没有x,y轴
# 准备数据

# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.pie(数据集, labels=数据集对应的分类)
plt.pie(moneys,labels=movies)
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")
#显示图例
plt.legend()
# 显示图像
plt.show()
```

```python
#直方图 一种柱状图，主要统计数据在各个区间中的分布情况，以柱子的高度表示数据在某区间中的数量，宽度表示区间的大小
# 准备数据
# 生成期望值为0，标准差为1的正态分布的若干数
random_numbers =  np.random.normal(0,1,10000)
#区间的个数
bins_number = 10
#使用列表来定义多个区间
bins_list = np.arange(11)-5 #生成[-5,5]的等分区间
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.hist(数据集, bins=区间设置) bins=n：将数据集最大值，最小值，坐等分n份， bins=列表，自定义的多个区间
plt.hist(random_numbers,bins=bins_list)
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")

# 显示图像
plt.show()
```

```python
#箱型图 检查异常值
# 准备数据
#生成1到10之间随机的500个数，分10行，5列
random_array = np.random.uniform(1,10,500).reshape(100,5)
#人为制造一个异常值
random_array[20,2] = 40
# 创建并配置画布
# figsize,dpi两个一起控制图片大小
figure = plt.figure(figsize=(10,6), dpi=100)
# 使用数据来绘制不同种类的图形
# plt.boxplot(二维数据集),按列检查数据集
plt.boxplot(random_array)
# 绘制图像的其他组件： 标题，x轴，y轴，刻度，图例，网格线
plt.title("hello,matplotlib!")

# 显示图像
plt.show()
```

```python
#画子图：一张画布上画多张图
#一次创建一个子图 plt.subplot()
#一次创建多个子图 plt.subplots()
# plt.subplots(nrows=几行，ncols=几列)
figure,axes = plt.subplots(nrows=2,ncols=2, figsize=(10,6),dpi=100)
#对0行0列的子图进行绘制
axes[0][0].plot(list1,list1)
#配置x轴，y轴相关信息
axes[0][0].set_xlabel("x轴的标签")
axes[0][0].set_xticks([1,3,5,7,9])
axes[0][0].set_xticklabels(["1度","3度","5度","7度","9度"])
axes[0][0].set_ylim([0,20]) # 设置y轴刻度范围
axes[0][0].legend(["图例1"])
axes[0][0].grid()
#对0行1列的子图进行绘制
axes[0][1].bar(movies,moneys)
#对1行0列的子图进行绘制
axes[1][0].pie(moneys,labels=movies)

#显示图形
plt.show()
```

```python
#案例：绘制美国某城市2014年7月的温度图
#准备数据
weather_df = pd.read_csv("./data/sitka_weather_07-2014.csv", usecols=["AKDT","Max_TemperatureF","Mean_TemperatureF","Min_TemperatureF"])
#创建画布
figure = plt.figure(figsize=(10,6),dpi=100)
plt.plot(weather_df["AKDT"],weather_df["Max_TemperatureF"],linestyle="--") #linestyle配置线形
plt.plot(weather_df["AKDT"],weather_df["Mean_TemperatureF"])
plt.plot(weather_df["AKDT"],weather_df["Min_TemperatureF"],linestyle="-.")
#绘制图形的其他组件
plt.title("美国某城市2014年7月的温度图", fontsize=24)
plt.ylabel("温度（华氏度）", fontsize=18)
plt.ylim([0,100])# 配置y轴刻度范围
#图例
plt.legend(["日最高温","日平均温","日最低温"])
#网格
plt.grid(alpha=0.3,linestyle="--") #alpha表示半透明度
#解决x轴显示重叠问题
figure.autofmt_xdate()
#保存图片,要在plt.show()前面保存
plt.savefig("./weather.png")
#显示图像，释放资源
plt.show()
```
